```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 9*x+8

x=np.linspace(0,3,300)
y=f(x)
noise=np.random.normal(0,1,300)
noised_y=y+noise
# print(noise)
# print(noised_y)
```

```python
from sklearn.linear_model import LinearRegression
x = np.array(x).reshape(-1,1)
lr = LinearRegression()
lr.fit(x, noised_y)
print(lr.coef_)
print(lr.intercept_)
plt.scatter(x, noised_y, label="Data")
plt.plot(x, lr.predict(x), color='red', label="Linear Regression")
```
